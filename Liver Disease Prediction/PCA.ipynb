```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Step 1: Load the dataset from Excel file
data = pd.read_csv('C:/Users/adhar/Jupiter notebook/Python/Data Analsis/Data-analysis/Liver Disease Prediction/Dataset/Liver Patient Dataset (LPD)_train.csv', encoding='ISO-8859-1')

# Display the first few rows of the dataset
print(data.head())

# Check for missing values and data types
print(data.info())
print(data['Result'].unique())

# Step 2: Remove rows with invalid 'Result' values (keep only 1 and 2)
data = data[data['Result'].isin([1, 2])]
print(data['Result'].unique())

# Step 3: Drop rows with missing values
data.dropna(inplace=True)

# Display the first few rows of the dataset after filtering
print(data.head())

# Check for missing values and data types
print(data.info())

# Step 4: Convert 'Result' values to binary (1 for liver disease, 0 for non-liver disease)
data['Liver Disease'] = data['Result'].map({1: 1, 2: 0})
print(data['Liver Disease'].unique())

# Step 5: Select relevant features for PCA
features = data[['Total Bilirubin', 'Direct Bilirubin', '\xa0Alkphos Alkaline Phosphotase',
                 '\xa0Sgpt Alamine Aminotransferase', 'Sgot Aspartate Aminotransferase', 'Total Protiens']] 

# ['Age of the patient', 'Gender of the patient', 'Total Bilirubin', 'Direct Bilirubin', '\xa0Alkphos Alkaline Phosphotase', 
# '\xa0Sgpt Alamine Aminotransferase', 'Sgot Aspartate Aminotransferase', 'Total Protiens', '\xa0ALB Albumin', 'A/G Ratio Albumin and Globulin Ratio', 
# 'Result']

# Step 6: Normalize the data (Min-Max Scaling)
normalized_features = (features - features.min()) / (features.max() - features.min())

# Step 7: Standardize the normalized data
scaler = StandardScaler()
scaled_features = scaler.fit_transform(normalized_features)

# Step 8: Apply PCA
pca = PCA(n_components=2)  # Reduce to 2 dimensions for visualization
principal_components = pca.fit_transform(scaled_features)

# Create a DataFrame with the principal components and the target variable
pca_df = pd.DataFrame(data=principal_components, columns=['Principal Component 1', 'Principal Component 2'])
pca_df['Liver Disease'] = data['Liver Disease']

# Step 9: Plotting the results
plt.figure(figsize=(10, 6))

# Scatter plot: red for liver disease (1), green for healthy (0)
colors = np.where(pca_df['Liver Disease'] == 1, 'red', 'green')
plt.scatter(pca_df['Principal Component 1'], pca_df['Principal Component 2'], c=colors, alpha=1)

plt.title('PCA of Liver Patients Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid()
plt.show()
```
