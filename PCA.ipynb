```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Load the dataset
liver_data = pd.read_csv("C:/Users/adhar/Downloads/ilpd_data.csv")

# Replace 'Female' with 0 and 'Male' with 1 in the GENDER column (if necessary)
'''liver_data['GENDER'] = liver_data['GENDER'].replace({'Female': 0, 'Male': 1})'''

# Step 1: Select 'SGPT' and 'TB' (Total Bilirubin) columns
X = liver_data[['SGPT', 'TB']]

# Step 2: Standardize the selected features
scaler = StandardScaler()
X_standardized = scaler.fit_transform(X)

# Step 3: Apply PCA
pca = PCA()
X_pca = pca.fit_transform(X_standardized)

# Step 4: Check the explained variance
explained_variance = pca.explained_variance_ratio_

# Output the principal components and explained variance
print("Principal Components (PCA Result):\n", X_pca)
print("\nExplained Variance by each Principal Component:\n", explained_variance)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Load the dataset
liver_data = pd.read_csv("C:/Users/adhar/Downloads/ilpd_data.csv")

# Replace 'Female' with 0 and 'Male' with 1 in the GENDER column (if necessary)
liver_data['GENDER'] = liver_data['GENDER'].replace({'Female': 0, 'Male': 1})

# Step 1: Select 'SGPT' and 'TB' (Total Bilirubin) columns
X = liver_data[['SGPT', 'TB']]

# Step 2: Standardize the selected features
scaler = StandardScaler()
X_standardized = scaler.fit_transform(X)

# Step 3: Apply PCA
pca = PCA()
X_pca = pca.fit_transform(X_standardized)

# Step 4: Check the explained variance
explained_variance = pca.explained_variance_ratio_

# Plot the PCA results (Principal Components)
plt.figure(figsize=(4,3))
plt.scatter(X_pca[:, 0], X_pca[:, 1], color='b', label='Principal Components')
plt.title('PCA of SGPT and Total Bilirubin')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.legend()
plt.show()

# Plot the explained variance
plt.figure(figsize=(4,3))
plt.bar(range(1, len(explained_variance) + 1), explained_variance, alpha=0.5, color='r', label='Explained Variance')
plt.xlabel('Principal Components')
plt.ylabel('Variance Explained')
plt.title('Explained Variance by Principal Components')
plt.legend()
plt.show()
```

```python
# Print the PCA components (loading matrix)
print("PCA Components (loading matrix):\n", pca.components_)
```
