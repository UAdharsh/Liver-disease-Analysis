```python
#Import Libraries
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
# Load Dataset
data = pd.read_csv('C:/Users/adhar/Jupiter notebook/Python/Data Analsis/Data-analysis/PCA/ilpd_data.csv')
```

```python
# Extracting features
X = data.drop(columns=['GENDER', 'SELECTOR'])  # Features
y = data['SELECTOR']  # Target variable

# Handle missing values (if any)
X = X.fillna(X.mean())
```

```python
# Standardize the Features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Apply PCA
pca = PCA(n_components=2)  # Reduce to 2 dimensions for visualization
X_pca = pca.fit_transform(X_scaled)
```

```python
# Visualize the Results
# Create a color mapping based on gender
colors = data['GENDER'].map({'Male': '#0096FF', 'Female': '#FF00FF'})

# Plotting the PCA results
plt.figure(figsize=(10, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=colors, alpha=0.7)
plt.xlabel('Principal Component 1') # variance related to liver function indicators such as SGPT and SGOT
plt.ylabel('Principal Component 2') # variance associated with other biochemical markers like TB (Total Bilirubin) and DB (Direct Bilirubin)
plt.title('PCA of Liver Patients Dataset')
plt.colorbar(label='Gender', ticks=[0, 1], format=plt.FuncFormatter(lambda x, _: 'Male' if x == '#0096FF' else 'Female'))
plt.grid()
plt.show()
```

```python
# Visualize the Results
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', alpha=0.7)
plt.xlabel('Principal Component 1') # variance related to liver function indicators such as SGPT and SGOT
plt.ylabel('Principal Component 2') # variance associated with other biochemical markers like TB (Total Bilirubin) and DB (Direct Bilirubin)
plt.title('PCA Result - Liver Patients Dataset')
plt.colorbar(label='Selector')
plt.grid()
plt.show()
```

```python
# Assuming 'colors' is defined based on SELECTOR
colors = data['SELECTOR'].map({1: 'red', 2: 'green'})  # Red for liver disease, Green for healthy

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=colors, alpha=0.7)
plt.xlabel('Principal Component 1') # variance related to liver function indicators such as SGPT and SGOT
plt.ylabel('Principal Component 2') # variance associated with other biochemical markers like TB (Total Bilirubin) and DB (Direct Bilirubin)
plt.title('PCA of Liver Patients Dataset')
plt.colorbar(label='Health Status', ticks=[1, 2], format=plt.FuncFormatter(lambda x, _: 'Liver Disease' if x == 1 else 'Healthy'))
plt.grid()
plt.show()
```

```python
# Evaluate Explained Variance
explained_variance_ratio = pca.explained_variance_ratio_
print("Explained Variance Ratio:", explained_variance_ratio)
```
